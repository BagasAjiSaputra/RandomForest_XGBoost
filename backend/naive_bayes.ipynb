## Membaca Dataset

```python
import pandas as pd
import pickle
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
import os
```

```python
df = pd.read_csv("dataset\WineQT.csv")
print("Dataset Loaded!")
df.head()
```

## Tipe Data

```python
df.info()
```

```python
df.quality.value_counts()
```

## Var Independen dan Dependen

```python
x = df.iloc[:, :-2]
y = df.iloc[:, -2]

y.head()
```

```python
x
```

## Split Data Testing dan Training

```python
x_train, x_test, y_train, y_test = train_test_split(x, y ,test_size=0.2, random_state=0, shuffle=False)
y_test
```

```python
x_test
```

```python
x_train
```

```python
y_train
```

## Training

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(x_train, y_train)

print("Training selesai!")
```

## Evaluasi

```python
from sklearn.metrics import ConfusionMatrixDisplay
y_pred = model.predict(x_test)

print("Akurasi:", accuracy_score(y_test, y_pred))
ConfusionMatrixDisplay.from_predictions(y_test, y_pred)
```

## Evaluasi Classification Report

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

## Evaluasi Imbalance Dataset (Wine)

```python
from sklearn.metrics import f1_score

print("F1 Macro:", f1_score(y_test, y_pred, average="macro"))
print("F1 Weighted:", f1_score(y_test, y_pred, average="weighted"))
```

## Simpan Data Evaluasi JSON

```python
import json
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score

metrics = {
    "accuracy": float(accuracy_score(y_test, y_pred)),
    "precision_macro": float(precision_score(y_test, y_pred, average="macro")),
    "recall_macro": float(recall_score(y_test, y_pred, average="macro")),
    "f1_macro": float(f1_score(y_test, y_pred, average="macro")),
    "f1_weighted": float(f1_score(y_test, y_pred, average="weighted"))
}

with open("Naive.json", "w") as f:
    json.dump(metrics, f, indent=4)

print("accuracy.json created!")
```
